```python
# anomaly_detection.ipynb

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import joblib

# ---------------- DEFINE BASE PATH ----------------
BASE_PATH = r"C:\Users\vishal\Desktop\web"  # <-- Change to your project folder
DATA_PATH = os.path.join(BASE_PATH, "data")
MODELS_PATH = os.path.join(BASE_PATH, "models")
PLOTS_PATH = os.path.join(BASE_PATH, "outputs", "visualization_plots")
REPORTS_PATH = os.path.join(BASE_PATH, "outputs", "reports")

# ---------------- ENSURE FOLDERS EXIST ----------------
os.makedirs(DATA_PATH, exist_ok=True)
os.makedirs(MODELS_PATH, exist_ok=True)
os.makedirs(PLOTS_PATH, exist_ok=True)
os.makedirs(REPORTS_PATH, exist_ok=True)

# ---------------- CHECK FILE ----------------
DATA_PATH = os.path.join(BASE_PATH, "data")
INPUT_FILE = os.path.join(DATA_PATH, "cleaned_dataset.csv")

def new_func(INPUT_FILE):
    raise FileNotFoundError(f"File not found: {INPUT_FILE}")

if not os.path.exists(INPUT_FILE):
    new_func(INPUT_FILE)

# ---------------- LOAD CLEANED DATA ----------------
df = pd.read_csv(INPUT_FILE)
print("✅ Loaded cleaned_dataset.csv")
print(df.head())

# ---------------- SELECT FEATURES ----------------
feature_cols = ["input_length", "num_special_chars", "num_digits", 
                "contains_sql_keyword", "contains_html_tag", "param_type"]
X = df[feature_cols]

# ---------------- TRAIN ISOLATION FOREST ----------------
iso_forest = IsolationForest(contamination=0.05, random_state=42)
df['anomaly_iforest'] = iso_forest.fit_predict(X)
df['anomaly_iforest'] = df['anomaly_iforest'].apply(lambda x: 1 if x == -1 else 0)
joblib.dump(iso_forest, os.path.join(MODELS_PATH, "isolation_forest_model.pkl"))
print("✅ Isolation Forest model saved.")

# ---------------- TRAIN ONE-CLASS SVM ----------------
oc_svm = OneClassSVM(nu=0.05, kernel="rbf", gamma='auto')
df['anomaly_ocsvm'] = oc_svm.fit_predict(X)
df['anomaly_ocsvm'] = df['anomaly_ocsvm'].apply(lambda x: 1 if x == -1 else 0)
joblib.dump(oc_svm, os.path.join(MODELS_PATH, "oneclass_svm_model.pkl"))
print("✅ One-Class SVM model saved.")

# ---------------- VISUALIZE ANOMALIES ----------------
plt.figure(figsize=(10,5))
sns.countplot(x='anomaly_iforest', data=df)
plt.title("Isolation Forest: Normal vs Anomalies")
plt.savefig(os.path.join(PLOTS_PATH, "isolation_forest_plot.png"))
plt.show()

plt.figure(figsize=(10,5))
sns.countplot(x='anomaly_ocsvm', data=df)
plt.title("One-Class SVM: Normal vs Anomalies")
plt.savefig(os.path.join(PLOTS_PATH, "oneclass_svm_plot.png"))
plt.show()

# ---------------- EVALUATION ----------------
if 'is_malicious' in df.columns:
    metrics = {}
    for model in ['anomaly_iforest','anomaly_ocsvm']:
        y_true = df['is_malicious']
        y_pred = df[model]
        metrics[model] = {
            "precision": precision_score(y_true, y_pred),
            "recall": recall_score(y_true, y_pred),
            "f1_score": f1_score(y_true, y_pred),
            "roc_auc": roc_auc_score(y_true, y_pred)
        }
    print("\nEvaluation Metrics:\n", metrics)

    # ROC comparison
    fpr1, tpr1, _ = roc_curve(df['is_malicious'], df['anomaly_iforest'])
    fpr2, tpr2, _ = roc_curve(df['is_malicious'], df['anomaly_ocsvm'])
    plt.figure(figsize=(8,6))
    plt.plot(fpr1, tpr1, label="Isolation Forest")
    plt.plot(fpr2, tpr2, label="One-Class SVM")
    plt.plot([0,1],[0,1],'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve Comparison")
    plt.legend()
    plt.savefig(os.path.join(PLOTS_PATH, "roc_comparison.png"))
    plt.show()

    # Save evaluation report
    with open(os.path.join(REPORTS_PATH,"evaluation_report.txt"),"w") as f:
        f.write(str(metrics))
    print("✅ Evaluation report saved.")

# ---------------- SAVE ANOMALY PREDICTIONS ----------------
OUTPUT_FILE = os.path.join(DATA_PATH, "anomaly_detection_results.csv")
df.to_csv(OUTPUT_FILE, index=False)
print("✅ Anomaly predictions saved to anomaly_detection_results.csv")
```
