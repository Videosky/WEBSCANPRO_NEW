```python
# data_preprocessing.ipynb

import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# ---------------- LOAD DATA ----------------
INPUT_FILE = "features_extracted.csv"
df = pd.read_csv(INPUT_FILE)
print("✅ Loaded features_extracted.csv")
print(df.head())

# ---------------- HANDLE MISSING DATA ----------------
print("\nMissing values per column:\n", df.isnull().sum())
# Fill missing input_value with empty string
if "input_value" in df.columns:
    df["input_value"] = df["input_value"].fillna("")
# Fill param_type if missing
if "param_type" in df.columns:
    df["param_type"] = df["param_type"].fillna("text")
# Remove rows with missing URL or parameter
df.dropna(subset=["url", "param_name"], inplace=True)

# ---------------- ENCODE CATEGORICAL ----------------
categorical_cols = ["url", "param_name", "param_type"]
le_dict = {}
for col in categorical_cols:
    if col in df.columns:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        le_dict[col] = le

# ---------------- NORMALIZE NUMERIC ----------------
numeric_cols = ["input_length", "num_special_chars", "num_digits"]
scaler = MinMaxScaler()
for col in numeric_cols:
    if col in df.columns:
        df[[col]] = scaler.fit_transform(df[[col]])

# ---------------- FINAL CLEANED DATA ----------------
OUTPUT_FILE = "cleaned_dataset.csv"
df.to_csv(OUTPUT_FILE, index=False)
print(f"\n✅ Cleaned dataset saved to {OUTPUT_FILE}")

# ---------------- VERIFY OUTPUT ----------------
print("\nFirst 5 rows:")
print(df.head())

print("\nDataset info:")
print(df.info())

print("\nDataset shape:")
print(df.shape)
```
