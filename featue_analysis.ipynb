```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("features_extracted.csv")
```

```python
sns.histplot(df["input_length"], bins=30, kde=True)
plt.show()
```

```python
sns.countplot(data=df, x="param_type")
plt.show()
```

```python
numeric_cols = ["input_length","num_special_chars","num_digits","contains_sql_keyword","contains_html_tag"]
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap="coolwarm")
plt.show()
```
