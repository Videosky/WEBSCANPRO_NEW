```python
# notebooks/feature_engineering_analysis.ipynb

# [Cell 1] - Improved Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
import warnings
warnings.filterwarnings('ignore')

# Set better visualization style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("🔧 Enhanced Feature Engineering Analysis")
print("=" * 50)

# [Cell 2] - Load the FIXED feature dataset
print("📥 Loading FIXED feature dataset...")
df = pd.read_csv(r"C:\Users\vishal\Desktop\WEBSCAN_PRO\data\feature_dataset.csv")

print(f"✅ Dataset shape: {df.shape}")
print(f"📊 Columns: {list(df.columns)}")
print("\nDataset Info:")
print(df.info())

# [Cell 3] - Enhanced Data Exploration
print("🔍 Enhanced Data Exploration")
print("\nFirst 5 rows:")
display(df.head())

print("\nDataset Description:")
print(df.describe())

print("\nMissing values:")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])

print("\nTarget distribution:")
print(f"Malicious (1): {df['is_malicious'].sum()} ({df['is_malicious'].mean()*100:.1f}%)")
print(f"Safe (0): {len(df) - df['is_malicious'].sum()} ({(1-df['is_malicious'].mean())*100:.1f}%)")

# [Cell 4] - Data Quality Assessment
print("🧹 Data Quality Assessment")

# Check for problematic values
zero_content = (df['html_content_length'] == 0).sum()
zero_errors = (df['error_message_flag'] == 0).sum()
total_records = len(df)

print(f"📊 Data Quality Metrics:")
print(f"  Total records: {total_records}")
print(f"  Zero content lengths: {zero_content} ({zero_content/total_records*100:.1f}%)")
print(f"  Records with no errors: {zero_errors} ({zero_errors/total_records*100:.1f}%)")
print(f"  Unique URLs: {df['url'].nunique()}")
print(f"  Unique status groups: {df['status_group'].nunique()}")

# Check if dataset needs cleaning
if zero_content > total_records * 0.5:
    print("⚠️  WARNING: High number of zero content lengths!")
else:
    print("✅ Content lengths look good!")

# [Cell 5] - Enhanced Feature Analysis
print("📈 Feature Distribution Analysis")

# Create subplots for feature distributions
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# 1. Response Time Distribution
safe_times = df[df['is_malicious'] == 0]['response_time']
malicious_times = df[df['is_malicious'] == 1]['response_time']

axes[0,0].hist(safe_times, alpha=0.7, label='Safe', bins=15, color='green')
axes[0,0].hist(malicious_times, alpha=0.7, label='Malicious', bins=15, color='red')
axes[0,0].set_title('Response Time Distribution')
axes[0,0].set_xlabel('Response Time (scaled)')
axes[0,0].set_ylabel('Frequency')
axes[0,0].legend()

# 2. HTML Content Length Distribution
safe_lengths = df[df['is_malicious'] == 0]['html_content_length']
malicious_lengths = df[df['is_malicious'] == 1]['html_content_length']

axes[0,1].hist(safe_lengths, alpha=0.7, label='Safe', bins=15, color='blue')
axes[0,1].hist(malicious_lengths, alpha=0.7, label='Malicious', bins=15, color='orange')
axes[0,1].set_title('HTML Content Length Distribution')
axes[0,1].set_xlabel('Content Length (scaled)')
axes[0,1].set_ylabel('Frequency')
axes[0,1].legend()

# 3. Error Flag Distribution by Class
error_dist = pd.crosstab(df['is_malicious'], df['error_message_flag'], normalize='index')
error_dist.plot(kind='bar', ax=axes[0,2], color=['lightcoral', 'coral'])
axes[0,2].set_title('Error Flag Distribution by Class')
axes[0,2].set_xlabel('Is Malicious')
axes[0,2].set_ylabel('Proportion')
axes[0,2].legend(['No Error', 'Error'])
axes[0,2].tick_params(axis='x', rotation=0)

# 4. Status Group Distribution
status_dist = pd.crosstab(df['is_malicious'], df['status_group'], normalize='index')
status_dist.plot(kind='bar', ax=axes[1,0], stacked=True)
axes[1,0].set_title('Status Group Distribution by Class')
axes[1,0].set_xlabel('Is Malicious')
axes[1,0].set_ylabel('Proportion')
axes[1,0].tick_params(axis='x', rotation=0)

# 5. Content Length Delta
safe_delta = df[df['is_malicious'] == 0]['content_length_delta']
malicious_delta = df[df['is_malicious'] == 1]['content_length_delta']

axes[1,1].boxplot([safe_delta, malicious_delta], labels=['Safe', 'Malicious'])
axes[1,1].set_title('Content Length Delta by Class')
axes[1,1].set_ylabel('Content Length Delta')

# 6. Payload Complexity
if 'payload_complexity' in df.columns:
    safe_complexity = df[df['is_malicious'] == 0]['payload_complexity']
    malicious_complexity = df[df['is_malicious'] == 1]['payload_complexity']
    
    axes[1,2].boxplot([safe_complexity, malicious_complexity], labels=['Safe', 'Malicious'])
    axes[1,2].set_title('Payload Complexity by Class')
    axes[1,2].set_ylabel('Payload Complexity')

plt.tight_layout()
plt.show()

# [Cell 6] - Statistical Comparison
print("📊 Statistical Comparison: Safe vs Malicious")

# Compare feature means between classes
feature_comparison = []
for feature in ['response_time', 'html_content_length', 'content_length_delta', 'response_time_normalized']:
    if feature in df.columns:
        safe_mean = df[df['is_malicious'] == 0][feature].mean()
        malicious_mean = df[df['is_malicious'] == 1][feature].mean()
        difference = malicious_mean - safe_mean
        feature_comparison.append({
            'Feature': feature,
            'Safe_Mean': safe_mean,
            'Malicious_Mean': malicious_mean,
            'Difference': difference,
            'Percent_Diff': (difference / safe_mean * 100) if safe_mean != 0 else 0
        })

comparison_df = pd.DataFrame(feature_comparison)
print("\nFeature Mean Comparison:")
display(comparison_df)

# [Cell 7] - Enhanced Correlation Analysis
print("🔗 Enhanced Correlation Analysis")

# Select numeric features for correlation
numeric_features = df.select_dtypes(include=[np.number])
# Remove URL if it was encoded
if 'url' in numeric_features.columns:
    numeric_features = numeric_features.drop('url', axis=1)

plt.figure(figsize=(12, 10))
correlation_matrix = numeric_features.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(correlation_matrix, 
            mask=mask,
            annot=True, 
            cmap='RdBu_r', 
            center=0,
            square=True, 
            fmt='.2f',
            cbar_kws={"shrink": .8})
plt.title('Feature Correlation Matrix (Lower Triangle)')
plt.tight_layout()
plt.show()

# Highlight high correlations with target
print("\n🎯 Correlation with Target (is_malicious):")
target_correlations = correlation_matrix['is_malicious'].sort_values(ascending=False)
print(target_correlations)

# [Cell 8] - Feature Importance Analysis
print("🎯 Feature Importance Analysis")

# Simple feature importance based on correlation
important_features = target_correlations[abs(target_correlations) > 0.1].drop('is_malicious')
print(f"Features with |correlation| > 0.1 with target:")
for feature, corr in important_features.items():
    print(f"  {feature}: {corr:.3f}")

# Visualize feature importance
plt.figure(figsize=(10, 6))
important_features.sort_values().plot(kind='barh', color='skyblue')
plt.title('Feature Correlation with Target (is_malicious)')
plt.xlabel('Correlation Coefficient')
plt.axvline(x=0, color='red', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# [Cell 9] - Advanced Feature Engineering (if needed)
print("🔧 Advanced Feature Engineering")

# Check if we need to create additional features
features_needed = []

# Check if payload features exist
if 'payload_length' not in df.columns:
    features_needed.append('payload_features')
    
if len(features_needed) > 0:
    print(f"Optional features that could be added: {features_needed}")
else:
    print("✅ All recommended features are present")

# [Cell 10] - Dataset Preparation for ML
print("🤖 Preparing Dataset for Machine Learning")

# Select features for modeling (excluding URL for now)
ml_features = [col for col in df.columns if col not in ['url', 'is_malicious'] and df[col].dtype in [np.number, bool]]

print(f"Selected {len(ml_features)} features for ML:")
print(ml_features)

# Create X and y for modeling
X = df[ml_features]
y = df['is_malicious']

print(f"\n🤖 ML Dataset Shape:")
print(f"X: {X.shape}")
print(f"y: {y.shape}")

# Check class balance
print(f"\n📊 Class Balance:")
print(f"Malicious: {y.sum()} ({y.mean()*100:.1f}%)")
print(f"Safe: {len(y) - y.sum()} ({(1-y.mean())*100:.1f}%)")

# [Cell 11] - Final Validation and Export
print("✅ Final Dataset Validation")

# Validation checks
validation_passed = True

# 1. Check for required features
required_features = ['response_time', 'html_content_length', 'error_message_flag', 'is_malicious']
for feature in required_features:
    if feature not in df.columns:
        print(f"❌ Missing required feature: {feature}")
        validation_passed = False

# 2. Check for feature variations
if df['html_content_length'].std() < 0.1:
    print("⚠️  Low variation in html_content_length")
    validation_passed = False

if df['error_message_flag'].sum() == 0:
    print("⚠️  No error flags detected")
    validation_passed = False

# 3. Check class balance
if y.mean() < 0.2 or y.mean() > 0.8:
    print(f"⚠️  Imbalanced dataset: {y.mean()*100:.1f}% malicious")

if validation_passed:
    print("🎉 Dataset validation PASSED! Ready for model training.")
    
    # Save prepared ML dataset
    ml_dataset = pd.concat([X, y], axis=1)
    ml_dataset_path = r"C:\Users\vishal\Desktop\WEBSCAN_PRO\data\ml_ready_dataset.csv"
    ml_dataset.to_csv(ml_dataset_path, index=False)
    print(f"💾 ML-ready dataset saved to: {ml_dataset_path}")
    
else:
    print("❌ Dataset validation FAILED! Check feature engineering.")

# [Cell 12] - Summary Report
print("\n" + "="*60)
print("📊 FEATURE ENGINEERING SUMMARY REPORT")
print("="*60)

print(f"📈 Dataset Overview:")
print(f"  Total Records: {len(df)}")
print(f"  Features: {len(df.columns)}")
print(f"  Malicious/Safe Ratio: {y.mean()*100:.1f}% / {(1-y.mean())*100:.1f}%")

print(f"\n🎯 Key Findings:")
print(f"  Strongest correlation with target: {target_correlations.index[1]} ({target_correlations.iloc[1]:.3f})")
print(f"  Error detection rate: {df['error_message_flag'].mean()*100:.1f}%")
print(f"  Content length variation: {df['html_content_length'].std():.3f}")

print(f"\n🚀 Next Steps:")
print(f"  1. Train classification models using ml_ready_dataset.csv")
print(f"  2. Evaluate model performance with cross-validation")
print(f"  3. Analyze feature importance from trained models")
print(f"  4. Deploy best-performing model for detection")

print("="*60)
```
